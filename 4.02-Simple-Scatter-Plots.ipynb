```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
```

## Точечные графики с `plt.plot`

В прошлом разделе мы рассматривали `plt.plot`/`ax.plot` для создания линейных графиков. Эта же функция может использоваться для создания точечных графиков:

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

Третий аргумент в функции (`'o'`) определяет стиль символа для отрисовка. Существует много различных символов:

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f"{marker=}")
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

Различные символы могут быть использованы вместе:

```python
plt.plot(x, y, '-ok')
```

Дополнительные аргументы `plt.plot()` задают еще более широкий диапазон настроек:

```python
plt.plot(
    x, y, '-p', color='gray',
    markersize=15,
    linewidth=4,
    markerfacecolor='white',
    markeredgecolor='gray',
    markeredgewidth=2
)
plt.ylim(-1.2, 1.2)
```

## Точечные графики с `plt.scatter`

Другой, более мощный, метод построения точечных графиков - это функция `plt.scatter`, которая используется очень похоже на `plt.plot`:

```python
plt.scatter(x, y, marker='o')
```

Главным различием между `plt.scatter` и `plt.plot` является то, что `plt.scatter` может быть использована для построения графиков где свойства каждой отдельной точки (размер, цвет заливки, цвет обводки и т.д.) могут быть индивидуально проконтролированы или заданы входными данными.

Давайте продемонстрируем это путем создания произвольного точечного графика с точками различных цветов и размеров. Мы также будем использовать прозрачность (`alpha` канал):

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()  # показать шкалу цветов
```

Обратите внимание, что аргумент `c` (цвета), автоматически сопоставлен на шкалу цвета и что аргумент размер задан в пикселях.

Например, мы можем использовать базу данных Iris из Scikit-Learn, где каждая записть является одним из трех типов цветов у которого лепестки и чашечки тщательно измерены:

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## `plot` vs `scatter`: заметка об эффективности

Для больших наборов данных (тысячи записей) `plt.plot` более эффективен чем `plt.scatter`.
